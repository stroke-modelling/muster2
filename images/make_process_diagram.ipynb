```python
import matplotlib.pyplot as plt
```

```python
# All

onset_to_call = 60

# Drip and ship

drip_ship = {
    'onset_to_call': 60,
    'call_to_ambo': 20,
    'ambo_on_scene': 30,
    'travel': 19,
    'door_to_needle': 30,
    'net_process_delay': 60,
    'transfer': 26,
    'door_to_puncture': 30
}

mothership = {
    'onset_to_call': 60,
    'call_to_ambo': 20,
    'ambo_on_scene': 30,
    'travel': 19,
    'door_to_needle': 30,
    'door_to_puncture': 45
}

mobile_stroke_unit = {
    'onset_to_call': 60,
    'msu_dispatch': 15,
    'travel_1': 35,
    'msu_ivt': 15,
    'post_ivt': 15,
    'travel_2': 35,
    'door_to_puncture': 45
}
```
